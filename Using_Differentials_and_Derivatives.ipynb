# Quadratic approximation

Let's have an arbitrary function $f(x)$. Suppose that we want to write $f$ as a quadratic polynomial such as

$$
f(x) = P(x) = a + b(x - x_0) + c(x - x_0)^2
$$

where, $x_0$ is called the pivot.

The first derivative of $P$ is

$$
P^\prime(x) = b + 2c(x - x_0)
$$

Evaluation of the first derivative at $x=x_0$ is

$$
P^\prime(x_0) = b
$$

The second derivative of $P$ is

$$
P^{\prime\prime}(x) = 2c
$$

As a result, $b=P^\prime(x_0)$ and $c=P^{\prime\prime}(x)/2$

Substitute back to the polynomial

$$
P(x) = a + P^\prime(x_0)(x - x_0) + \frac{P^{\prime\prime}(x)}{2}(x - x_0)^2
$$

# General case of approximation (Taylor series)

In general, a function can be approximated as a polynomial around $x_o$ with

$$
f(x) = \sum_{n=0}^{\infty} \frac{f^{(n)}(x_0)}{n!}(x - x_0)^n
$$

Keep in mind that not all functions can be represented as polynomials. So, a function may not have Taylor series.

## Example 1: Change of area of a circle with radius.

Area of a circle is

$$
A = \pi r^2
$$

The following figure shows area (the blue curve) as a function of radius.



```python
import e1
```

Let's say the initial radius is $r_0=2$. Then, initial area is $A_0 = 4\pi$.

100% increase in radius means

$$
r = r_0 + \frac{150}{100}r_0 = r_0 + 1.5r_0 = 2 + 1.5(2) = 5
$$

The corresponding area is $A = 25\pi$ (see the figure).

### Linear approximation

Let's approximate area linearly (using only the first derivative) around the pivot $r_0$. Slope of the tangent line (red) or first derivative of area at $r_0$ is

$$
A^\prime(r=2) = \left.\frac{dA}{dr}\right|_{r=2} = 2\pi r\rvert_{r=2} = 4\pi
$$

Linearly approximated area is

$$
A_{a1} = A_0 + \left.\frac{dA}{dr}\right|_{r=2}(r - r_0)
$$

$$
A_{a1} = 4\pi + 4\pi (3) = 16\pi
$$

The difference between exact area and linear approximation is $\text{error} = (25 - 16)\pi = 9\pi$

### Percent change

Fractional change in radius is

$$
\frac{\Delta r}{r_0} = \frac{r - r_0}{r_0} = \frac{5 - 2}{2} = \frac{3}{2} = \frac{150}{100}
$$

Fractional change in area is

$$
\frac{\Delta A}{A_0} = \frac{A - A_0}{A_0} = \frac{A^\prime(r_0) \Delta r}{A} = \frac{2\pi r_0 \Delta r}{\pi r_0^2} = 2\frac{\Delta r}{r_0}
$$

Substitution of fractional change in radius yields

$$
\frac{\Delta A}{A_0} = 2\frac{150}{100} = \frac{300}{100}
$$

Therefore, when radius increases 150%, area increases 300%. In general, the change in area is double the change in radius.

### Second order approximation

$$
A_{a2} = A_0 + \left.\frac{dA}{dr}\right|_{r=2}\Delta r + \frac{1}{2}\left.\frac{d^2A}{dr^2}\right|_{r=2}(\Delta r)^2
$$

or

$$
A_{a2} = A_{a1} + \frac{1}{2}\left.\frac{d^2A}{dr^2}\right|_{r=2}(\Delta r)^2
$$

Second derivative of $A$ is

$$
A^{\prime\prime} = \frac{d^2A}{dr^2} = 2\pi
$$

regardless of $r$. Substituting all values,

$$
A_{a2} = 16\pi + \frac{1}{2}2\pi(3)^2 = 16\pi + 9\pi = 25\pi
$$

The approximated area is equal to the exact area. This is not suprising because area ($A=\pi r^2$) is of second order or quadratic or parabolic. Therefore, second order (quadratic) approximation is enough to obtain the exact value.

## Example 2: Find the approximation of the following function at $t=0$ using $t_0=-5$ as pivot.

$$
T = \frac{1}{3}t^3 - 3t^2 + 8t + 10
$$

For the pivot $t_0=-5$, corresponding function output is

$$
T_0 = T(-5) = \frac{1}{3}(-5)^3 - 3(-5)^2 + 8(-5) + 10 = 146.7
$$

The exact value of the function is $T(t_0=0) = 0$.

```python
import e2
```

### Linear approximation

First derivative of $T$ is

$$
\frac{dT}{dt} = t^2 - 6t + 8
$$

Evaluation of the first derivative at the pivot $t_0=-5$ is

$$
\left.\frac{dT}{dt}\right|_{t=-5} = (-5)^2 - 6(-5) + 8 = 63
$$

Linear approximation of $T(0)$ is

$$
T_{a1} = T_0 + \left.\frac{dT}{dt}\right|_{t=-5} \Delta t
$$

$$
T_{a1} = 10 + 63[0-(-5)] = 168.3
$$

Magnitude of error is $|63 - 163.3| = 100.3$.

### Quadratic approximation

Second derivative of $T$ is

$$
\frac{d^2T}{dt^2} = 2t - 6
$$

Evaluation of the second derivative at the pivot $t_0=-5$ is

$$
\left.\frac{d^2T}{dt^2}\right|_{t=-5} = 2(-5) - 6 = -16
$$

Second order approximation is

$$
T_{a2} = T_0 + \left.\frac{dT}{dt}\right|_{t=-5} \Delta t + \frac{1}{2}\left.\frac{d^2T}{dt^2}\right|_{t=-5} (\Delta t)^2
$$

or

$$
T_{a2} = T_{a1} + \frac{1}{2}\left.\frac{d^2T}{dt^2}\right|_{t=-5} (\Delta t)^2
$$

$$
T_{a2} = 168.33 + \frac{1}{2} (-16) [0 - (-5)]^2 = -31.7
$$

Magnitude of error is $|63 - -31.7| = 94.7$. It is getting better.

### Cubic approximation

Third derivative of $T$ is

$$
\frac{d^2T}{dt^2} = 2
$$

Third order approximation is

$$
T_{a2} = T_0 + \left.\frac{dT}{dt}\right|_{t=-5} \Delta t + \frac{1}{2}\left.\frac{d^2T}{dt^2}\right|_{t=-5} (\Delta t)^2 + \frac{1}{6}\left.\frac{d^3T}{dt^3}\right|_{t=-5} (\Delta t)^3
$$

or

$$
T_{a3} = T_{a2} + 2 [0 - (-5)]^3 = 10
$$

As expected, third order approximation results in exact answer for a cubic function.
